```python
# import libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
base_url = "https://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&SIC="
loop_url = "&owner=include&match=&start="
end_url = "&count=40&hidefilings=0"
```

```python
finance = [6172, 6199]
banks = [6021, 6022, 6029, 6099, 6162]
insurances = [6311, 6321, 6331, 6351, 6361, 6399, 6411]
loans = [6162, 6163]

all_finance_SIC = finance + banks + insurances + loans

all_finance_SIC
```

```python
def getCIK(SIC_number):
    CIK = []
    url = []
    company_name = []

    # we try one SIC first
    test_sic = 6199

    # start the page with 0
    base_loop_count = 0

    while True:
        # construct the scraping url
        scraped_url = base_url + str(SIC_number) + loop_url + str(base_loop_count) + end_url
        print(scraped_url)
        url_response = requests.get(scraped_url)
        url_html = BeautifulSoup(url_response.content, "html.parser")

        if url_html.find_all('div', attrs={"id": "contentDiv"})[0].text != '\nNo matching companies.\n':
            print('ok')
            tr_test = url_html.find_all('tr')
            for rows in tr_test[1:]:
                ind_row = rows.find_all('td')
                CIK.append(ind_row[0].a.text)
                url.append(ind_row[0].a['href'])
                company_name.append(ind_row[1].text)
        else:
            break
        base_loop_count += 40

    CIK_full = pd.DataFrame({"CIK": CIK, "url": url, "company": company_name})

    return CIK_full
```

```python
getCIK(6199)
```

```python
SIC_df = []

for sic in all_finance_SIC:
    df = getCIK(sic)
    SIC_df.append(df)
```

```python
for i in SIC_df:
    print(len(i))
```

```python
combined_CIK = pd.concat(SIC_df)

len(combined_CIK)
```
