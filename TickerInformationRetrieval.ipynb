```python
import time
import datetime
df_list = []

for i in tickers_till_AOG:
    #alphavantage link
    url = str(base_url + i + base_url_end + API_key)
    ticker_data = str(i)
    
    #parsing
    test_1 = requests.get(url)
    parsed_data_1 = test_1.json()
    parsed_data_1
    
    open_list_1 = []
    close_list_1 = []
    vol_list_1 = [] 
    
    open_list_2 = []
    close_list_2 = []
    vol_list_2 = []
    
    #create dataframe for ticker
    ticker_df = df[df['Ticker']==ticker_data]
    
    for date in news_df['realfiledDate']:
        first_day = datetime.datetime.strptime(date, "%Y-%m-%d")
        if first_day.weekday() == 4:
            next_day = first_day + datetime.timedelta(days=3)
            next_day_str = next_day.strftime("%Y-%m-%d")
        else:
            next_day = first_day + datetime.timedelta(days=1)
            next_day_str = next_day.strftime("%Y-%m-%d")
        try:
            #first day
            open_list_1.append(parsed_data_1['Time Series (Daily)'][date]["1. open"])
            close_list_1.append(parsed_data_1['Time Series (Daily)'][date]["4. close"])
            vol_list_1.append(parsed_data_1['Time Series (Daily)'][date]["5. volume"])
            
            open_list_2.append(parsed_data_1['Time Series (Daily)'][next_day_str]["1. open"])
            close_list_2.append(parsed_data_1['Time Series (Daily)'][next_day_str]["4. close"])
            vol_list_2.append(parsed_data_1['Time Series (Daily)'][next_day_str]["5. volume"])
            
            print("ticker {ticker} part done for {date}".format(ticker=ticker_data,date=date))
        except:
            open_list_1.append("NA")
            close_list_1.append("NA")
            vol_list_1.append("NA")
            
            open_list_2.append("NA")
            close_list_2.append("NA")
            vol_list_2.append("NA")
            
            print("Something went wrong with {ticker} part - date {date}".format(ticker=ticker_data,date=date))
    
    #ocv data to dict
    print("converting")
    dict_ocv_data = {"open_1":open_list_1,"close_1":close_list_1,"vol_1":vol_list_1,"open_2":open_list_2,"close_2":close_list_2,"vol_2":vol_list_2}
    #Dict to Df
    print(len(open_list_1))
    print(len(close_list_1))
    print(len(vol_list_1))
    print(len(open_list_2))
    print(len(close_list_2))
    print(len(vol_list_2))
    print("converting")
    df_ocv_data = pd.DataFrame(dict_ocv_data)
    #combine them together
    print("joining")
    df_new_ticker = ticker_df.join(df_ocv_data)
    print("appending")
    df_list.append(df_new_ticker)
    print("Ticker {ticker} set done".format(ticker=ticker_data))
    time.sleep(61)
    
```

```python
concat_df = pd.concat(df_list)
concat_df.to_csv("df_extracted_ocv_till_AOC.csv")
```
