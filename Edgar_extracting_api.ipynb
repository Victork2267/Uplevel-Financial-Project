## Extracting open close and vol

```python
import time
# Contain day 1 data 
open_data_f = []
close_data_f = []
vol_data_f = []

# Contain next day data
open_data_c = []
close_data_c = []
vol_data_c = []

callcounter = 0

#loop through ticker and dates
for ticker,date in zip(df["Ticker"],df["realfiledDate"]):
    #combine the ticker with the url
    av_url = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=" + str(ticker) + "&outputsize=full&apikey=MDLC9RXAATFB41HB"
    #load up in json
    with urllib.request.urlopen(av_url) as url:
        data = json.loads(url.read().decode())
    #loop through the dates
    try:
        for i in data["Time Series (Daily)"]:
        #check if the date is inside the json file and extract and append
            if i == date:
                open_data_1 = data["Time Series (Daily)"][i]["1. open"]
                open_data_f.append(open_data_1)
                close_data_1 = data["Time Series (Daily)"][i]["4. close"]
                close_data_f.append(close_data_1)
                vol_data_1 = data["Time Series (Daily)"][i]["5. volume"]
                vol_data_f.append(vol_data_1)
                # next day data
                first_day = datetime.datetime.strptime(date, "%Y-%m-%d")
                next_day = first_day + datetime.timedelta(days=1)
                next_day_str = next_day.strftime("%Y-%m-%d")
                # extract and append
                open_data_2 = data["Time Series (Daily)"][next_day_str]["1. open"]
                open_data_c.append(open_data_2)
                close_data_2 = data["Time Series (Daily)"][next_day_str]["4. close"]
                close_data_c.append(close_data_2)
                vol_data_2 = data["Time Series (Daily)"][next_day_str]["5. volume"]
                vol_data_c.append(vol_data_2)
    except:
        no_text = "NA"
        open_data_f.append(no_text)
        close_data_f.append(no_text)
        vol_data_f.append(no_text)
        
        open_data_c.append(no_text)
        close_data_c.append(no_text)
        vol_data_c.append(no_text)
    
    if callcounter < 6:
        callcounter += 1
    else:
        time.sleep(61)
        callcounter = 0
        
    print("row done")
```

```python
stock_value = {"open_first":open_data_f,
              "close_first":close_data_f,
              "vol_first":vol_data_f,
              "open_second":open_data_c,
              "close_second":close_data_c,
              "vol_second":vol_data_c}

df_stock_data = pd.DataFrame(stock_value)
df_stock_data.head(20)
```
